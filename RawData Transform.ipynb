```python
Name = []
DatePublished = []
Description = []
Url = []


for i in df_json:

    article = json.loads(i)

    try:
        if article.get('json_object').get('datePublished'):

            Name.append(article['json_object']['name'])
            DatePublished.append(article['json_object']['datePublished'])
            Description.append(article['json_object']['description'])
            Url.append(article['json_object']['url'])
    except Exception as e:
        print(f" Error processing : {e}")
```

```python
from pyspark.sql.types import *

datadef = list(zip(Name,DatePublished,Description,Url))

schemadef = StructType([
    StructField("name",StringType(),True),
    StructField("datePublished",StringType(),True),
    StructField("description",StringType(),True),
    StructField("url",StringType(),True)
])


dataset = spark.createDataFrame(data=datadef, schema=schemadef)
```

```python
from pyspark.sql.functions import *
dataset = dataset.withColumn('dateformat', date_format("DatePublished","yyyy-MM-dd"))
dataset = dataset.withColumn('Published', to_date('dateformat', "yyyy-MM-dd"))
dataset = dataset.drop('dateformat',"DatePublished")
display(dataset)
```

```python
from pyspark.sql.utils import *


#dataset.write.format('delta').saveAsTable('cleandata')


try:
    table_name = 'cleandata'

    dataset.write.format('delta').saveAsTable(table_name)


except AnalysisException:

    print('Table Already Exists')
    dataset.createOrReplaceTempView('vw_dataset')

    spark.sql(f""" MERGE INTO {table_name} target
                   USING vw_dataset source
                   ON target.Url = source.Url

                   WHEN MATCHED AND
                   target.Name <> source.Name OR
                   target.Description <> source.Description OR
                   target.Published <> source.Published

                   THEN UPDATE SET *

                   WHEN NOT MATCHED 
                   THEN INSERT * 
    
     """)

```
